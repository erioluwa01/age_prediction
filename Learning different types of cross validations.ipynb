```python
import pandas as pd
import numpy as np
df=pd.read_csv('Cancer_Data.csv')
```

```python
df.head(30)
```

```python
#I selected my independent and dependent feature
x=df.iloc[:, 2:]
y=df.iloc[:,1]
```

```python
x.head()
```

```python
x=x.dropna(axis=1)
```

```python
x.head()
```

```python
#Now lets check out the various types of cross validation
```



# HoldOut Validation Approch- Train_test_split

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
predict=model.predict(x_test)
print(predict)
```

### This calculates the accuracy of your trained DecisionTreeClassifier on the test data.
#It compares your model’s predictions (predict) with the actual values (y_test).
Formula for Accuracy:
Accuracy=
Correct Predictions/
Total Predictions
Accuracy= 
Total Predictions
Correct Predictions




```python
result = model.score(x_test, y_test)
print(result)
```

# Now let's look at K Fold Cross Validation

```python
# Here I can make k number of splits and measure the accuracy of each of those categories
```

```python
from sklearn.model_selection import KFold
model = DecisionTreeClassifier()
kfold_validation=KFold(10)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
results=cross_val_score(model, x,y, cv=kfold_validation)
print(results)
```

```python
# so let's say I want to get the mean of my result
print(np.mean(results))
```

### So, basically this results is showing the accuracy for each of the 10 segments the data has being splitted into

# Sratified K Fold Cross Validation
##THis is the method we implore when wr have unbalanced dataset

```python
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
scores=cross_val_score(model, x,y, cv=skfold)
```

```python
scores
```

```python
print(np.mean(scores))
```

# Leave One Out Cross Validation (LOOCV)

#The disadvantages of this is that it depends on the value of random state, because bothe the test and train value are selcted randomly, also if I'm dealing with a large data set I'd have to do multiple iterations..
 


```python
from sklearn.model_selection import LeaveOneOut
model=DecisionTreeClassifier()
leave_validation=LeaveOneOut()
results=cross_val_score(model, x,y,cv=leave_validation)
```

```python
results
```

```python
print(np.mean(results))
```
